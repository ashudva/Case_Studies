```r
options(digits = 3)
library(titanic)
library(tidyverse)

titanic <- titanic_train %>%
    select(Survived, Pclass, Sex, Age, SibSp, Parch, Fare) %>%
    mutate(Survived = factor(Survived),
           Pclass = factor(Pclass),
           Sex = factor(Sex))
```

```r
str(titanic)
```

```r
head(titanic)
```

```r
titanic %>%
    filter(!is.na(Age)) %>%
    ggplot(aes(Age, color = Sex)) +
    geom_density()
```

```r
titanic %>%
    filter(!is.na(Age)) %>%
    ggplot(aes(Age, fill = Sex)) +
    geom_density(alpha = 0.2) +
    facet_grid(.~Sex)
```

```r
male <- titanic %>%
    select(Sex) %>%
    filter(Sex == "male")
female <- titanic %>%
    select(Sex) %>%
    filter(Sex == "female")
str(female)
```

```r
titanic %>%
    filter(!is.na(Age)) %>%
    ggplot(aes(Age, fill = Sex)) +
    geom_histogram() +
    facet_grid(.~Sex)
```

```r
summary(titanic)
```

```r
params <- titanic %>%
    filter(!is.na(Age)) %>%
    summarize(mean = mean(Age), sd = sd(Age))

titanic %>%
    ggplot(aes())
```
